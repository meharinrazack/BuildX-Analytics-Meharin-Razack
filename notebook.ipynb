```python
df.rename(columns={
    'restaurant name':'restaurant_name',
    'restaurant type':'restaurant_type',
    'rate (out of 5)':'rating',
    'num of ratings':'num_ratings',
    'avg cost (two people)':'cost_for_two',
    'cuisines type':'cuisine_type',
    'local address':'local_address'
}, inplace=True)
# Renaming columns for consistency and easier coding
```

```python
df.drop(
    columns=['Unnamed: 0.1','Unnamed: 0'],
    inplace=True
)
# Removing unwanted index columns created during dataset export
```

```python
# Filling missing values in rating column with median
# Median is used because it is less affected by extreme values

df['rating'] = df['rating'].fillna(
    df['rating'].median()
)

# Filling missing values in cost_for_two column with median
# This preserves all restaurant records for analysis

df['cost_for_two'] = df['cost_for_two'].fillna(
    df['cost_for_two'].median()
)
df[['rating', 'cost_for_two']].isnull().sum()
```

 Missing values were replaced using median values.
 Median is preferred over mean because it is less sensitive to outliers.
 This allows us to keep all restaurant records instead of deleting rows.

```python
df.drop_duplicates(inplace=True)
# Removing duplicate restaurant records
```

```python
text_columns = [
    'restaurant_name',
    'restaurant_type',
    'cuisine_type',
    'area',
    'local_address'
]

for col in text_columns:
    df[col] = df[col].str.strip().str.title()
    # Cleaning text fields by removing extra spaces
# and standardizing capitalization
```

```python
df.dtypes
df['rating'] = pd.to_numeric(
    df['rating'],
    errors='coerce'
)

df['num_ratings'] = pd.to_numeric(
    df['num_ratings'],
    errors='coerce'
)

df['cost_for_two'] = pd.to_numeric(
    df['cost_for_two'],
    errors='coerce'
)
# Ensuring numerical columns have correct data types
```

```python
df['cost_per_person'] = (
    df['cost_for_two'] / 2
)
# Calculating approximate cost per person
```

```python
df['rating_category'] = pd.cut(
    df['rating'],
    bins=[0,3,4,5],
    labels=['Low','Average','Excellent']
)
# Categorizing restaurants based on ratings
```
